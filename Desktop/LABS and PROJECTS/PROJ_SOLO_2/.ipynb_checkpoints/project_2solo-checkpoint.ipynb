```python
def percent_success(df):
    # Sort by user by only first start
    df = df.sort_values(by=['client_id', 'date_time'])
    
    first_start = df[df['process_step'] == 'start'].drop_duplicates(subset=['client_id'], keep='first') # Filter only the first 'start' for each user

    
    start_count = first_start['client_id'].nunique()  # count the number of unique users who started the process

    # find users who have 'confirm' step
    confirmed_users = df[df['process_step'] == 'confirm']['client_id'].unique()

    confirm_count = first_start[first_start['client_id'].isin(confirmed_users)]['client_id'].nunique()
    success_rate = (confirm_count / start_count) * 100 if start_count > 0 else 0

    print(f"The overall success rate for completing the process is: {success_rate:.2f}%")

    return success_rate
```

```python
percent_success(df_control_group_clinfo) 

percent_success(df_test_group_clinfo)    
```

# completion rate for 50-70 group

```python
def percent_success_age(df):
    # Filter the dataframe for clients in the age group between 50 and 70
    age_filtered_df = df[(df['clnt_age'] >= 50) & (df['clnt_age'] <= 70)]
    
   
    age_filtered_df = age_filtered_df.sort_values(by=['client_id', 'date_time'])

    first_start = age_filtered_df[age_filtered_df['process_step'] == 'start'].drop_duplicates(subset=['client_id'], keep='first')

   
    start_count = first_start['client_id'].nunique()

    
    confirmed_users = age_filtered_df[age_filtered_df['process_step'] == 'confirm']['client_id'].unique()

    confirm_count = first_start[first_start['client_id'].isin(confirmed_users)]['client_id'].nunique()

    success_rate = (confirm_count / start_count) * 100 if start_count > 0 else 0

    print(f"The overall success rate for completing the process is: {success_rate:.2f}%")

    return success_rate
```

```python
percent_success_age(df_control_group_clinfo) 

percent_success_age(df_test_group_clinfo)    
```
